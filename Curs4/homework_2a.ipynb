### Task 3

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y1, y2, y3, y4
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim([-5,5])
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
url = 'creditcard.csv'
credit = pd.read_csv(url)
```

```python
credit.head()
```

```python
df = pd.DataFrame(credit)
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.index
```

```python
df_class = df['Class'].value_counts()
df_class
```

```python
df_class.plot(kind='bar')
plt.yscale(value='log')
plt.show()
```

```python
v11 = df.query('Class == 1')['V1']
v1Class1 = v11
v1Class1
```

```python
v10 = df.query('Class == 0')['V1']
v1Class0 = v10
v1Class0
```

```python
plt.hist(v1Class1, bins = 20, color = 'red')
plt.hist(v1Class0, bins = 20, color = 'grey', alpha = 0.5)
plt.yscale(value='log')
plt.xlabel('V1')
plt.legend(labels = 'C')

plt.show()
```
