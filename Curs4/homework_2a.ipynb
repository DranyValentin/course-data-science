### Task 2

```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color='green')
plt.axis([0.5,9.5,-2.5,2.5])
plt.title(' f(t)')
plt.xlabel(' t')
plt.ylabel(' f')
plt.show()
```

### Task 3

```python
x = np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y1, y2, y3, y4
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim([-5,5])
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```
