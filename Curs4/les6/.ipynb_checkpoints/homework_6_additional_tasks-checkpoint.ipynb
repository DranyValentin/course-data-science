### Корреляция данных (нормализация данных)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X.head()
```

```python
keys = X.keys()
X_corr = pd.DataFrame(scaler.fit_transform(X), columns = keys)
X_corr.head()
```

```python
X_corr.describe()
```

```python
high_corr = X_corr.loc[X_corr['target'] >= 0.5, keys.drop('target')]
high_corr
#X_corr.loc[pd.Series([True, True, True, True, True, True, True, True, True, True, True, True, True, False],
#       index=['viper', 'sidewinder', 'cobra'])]
```

```python
(high_corr ** 2)
```

```python
high_corr2 = high_corr.rename(columns = lambda x: x + '_2')
high_corr2
```

```python
X = X.drop('target', axis = 1)
X = pd.concat([X, high_corr2 ** 2], axis = 1, ignore_index = False)
X
```
