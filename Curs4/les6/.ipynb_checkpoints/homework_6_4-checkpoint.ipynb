```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
creditcard_data = pd.read_csv('creditcard.csv')
creditcard_data
```

```python
creditcard_data.shape
```

```python
df_creditcardData = pd.DataFrame(creditcard_data)
df_creditcardData.head()
```

```python
df_creditcardData.value_counts(normalize = True)
```

```python
df_creditcardData.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df_creditcardData.head(10)
```

```python
X = df_creditcardData.drop(['Class'], axis=1)
X
```

```python
df_creditcardData.describe()
```

```python
y = pd.Series(df_creditcardData['Class'], copy = True)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100, stratify = y)
```

```python
X_train.shape
```

```python
parameters = [{'n_estimators': [10, 15],
              'max_features': np.arange(3,5),
              'max_depth': np.arange(4,7)}]
```

```python
clf = GridSearchCV(estimator = RandomForestClassifier(random_state = 100),
                  param_grid = parameters,
                  scoring = 'roc_auc',
                  cv = 3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
clf = RandomForestClassifier(max_depth = 6, max_features = 3, n_estimators = 15, random_state = 100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
y_pred_proba = clf.predict_proba(X_test)
y_pred_proba = y_pred_proba[:, 1]
```

```python
from sklearn.metrics import roc_curve
from pylab import rcParams
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label = 1)
```

```python
rcParams['figure.figsize'] = 5, 4.5
plt.plot(fpr, tpr, color = 'green', label = 'ROC curve')
plt.plot([0,1], [0,1], color='grey', linestyle = '--')
plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Postive Rate')
plt.title('Receiver operating characteristic (ROC)')
```
